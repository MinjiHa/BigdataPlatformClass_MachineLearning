```python
import seaborn as sns
anscombe = sns.load_dataset("anscombe")
anscombe # 통계치로만 데이터를 해석하지 말아라!! 를 나타내주는 dataset

anscombe.describe() #기술통계를 나타낼 때 discribe()
# dataset 단위로 x 및 y의 평균 출력
anscombe.groupby('dataset').mean() #평균이 같게 나옴
anscombe.groupby('dataset').std() #표준편차 같게 나옴
anscombe.groupby('dataset').var() #분산 같게 나옴
```

```python
person = pd.read_csv("../data2/survey_person.csv")
print(person)
print("="*50)
site = pd.read_csv("../data2/survey_site.csv")
print(site)
print("="*50)
survey = pd.read_csv("../data2/survey_survey.csv")
print(survey)
print("="*50)
visited = pd.read_csv("../data2/survey_visited.csv")
print(visited)
print("="*50)
```

```python
visitedSubset = visited.loc[[0,2,6]]
visitedSubset
```

```python
site
```

```python
o2oMerge = site.merge(visitedSubset, left_on='name', right_on='site')
o2oMerge
```

```python
# site의 name과 visited의 site 기준으로 merge
m2oMerge = site.merge(visited, left_on='name', right_on='site')
m2oMerge

# person의 ident과 survey의 person 기준으로 merge
ps = person.merge(survey, left_on='ident', right_on='person')

# visited의 ident과 survey의 taken 기준으로 merge
vs = visited.merge(survey, left_on='ident', right_on='taken')
vs
```

```python
print(ps)
print("="*60)
print(vs)

psvs = ps.merge(vs, left_on= ['ident','taken','quant','reading'] , right_on=['person','ident','quant','reading'] )
psvs
```
